```python
#7.Check if a number is a palindrome. 
n=int(input("Enter the no."))
reverse=0
original=n

while n>0:
    digit=n%10
    reverse=reverse*10+digit
    n//=10

if original == reverse:
    print("Palindrome number")
else:
    print("Not a palindrome number")
```

```python
#10. Write a program to print all prime numbers between 1 and 100. 

for num in range(2, 101):
    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num)
```

```python
for i in range(1, 1001):
    num = i
    digits = 0

    # Count digits
    temp = num
    while temp > 0:
        digits += 1
        temp //= 10

    # Armstrong calculation
    temp = num
    armstrong_sum = 0
    while temp > 0:
        n = temp % 10
        armstrong_sum += n ** digits
        temp //= 10

    if armstrong_sum == num:
        print(num)
```

```python
#Write a function to count the frequency of elements in a list.
def frequency_no(ls):
    freq={}
    for item in ls:
        if item in freq:
            freq[item] +=1
        else:
            freq[item] =1 
    return freq
    

my_list = [1, 2, 2, 3, 4, 3, 2, 1, 5]
print(frequency_no(my_list))
```

```python
# Write a function to find common elements in two lists. 


def common(l1,l2):
    comm=[]
    for i in l1:
        for j in l2:
            if i==j:
                comm.append(i)
    return comm

list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
print(common(list1,list2))
```

```python
#Write a function to remove all punctuation from a string. 

import string

def remove_punc(word):
    new=""
    for i in word:
        if i not in string.punctuation:
            new +=i
    return new

sentence = "Hello, world! Welcome to Python."
print(f"New setence is : {remove_punc(sentence)}")
```

```python
#Create a function to rotate a list left by k positions. 

# Create a function to rotate a list left by k positions

def rotated(l1):
    new = []
    k = int(input("Enter position from where you want to shift: "))
    n = len(l1)
    
    # Ensure k is within bounds
    k = k % n

    # Append elements from k to end
    for i in range(k, n):
        new.append(l1[i])
    
    # Append elements from start to k
    for i in range(k):
        new.append(l1[i])
    
    return new

# Example usage
numbers = [1, 2, 3, 4, 5]
rotated_list = rotated(numbers)
print("Rotated List:", rotated_list)
```

```python
#Write a function to merge two dictionaries and handle key collisions by summing values. 

def merge_dicts_sum(dict1, dict2):
    merged = dict1.copy()  # start with dict1
    for key, value in dict2.items():
        if key in merged:
            merged[key] += value  # sum values if key exists
        else:
            merged[key] = value   # add new key
    return merged

# Example usage
dict1 = {'apple': 10, 'banana': 5, 'orange': 8}
dict2 = {'banana': 3, 'orange': 7, 'grapes': 6}

merged_dict = merge_dicts_sum(dict1, dict2)
print("Merged Dictionary:", merged_dict)
```

```python
# Create a function to find unique elements present in only one of two lists. 

def unique_elements(l1,l2):
    new=[]
    for i in l1:
        if i not in l2:
            new.append(i)
    return new
            


list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]

result = unique_elements(list1, list2)
print("Unique elements in only one list:", result)
```
